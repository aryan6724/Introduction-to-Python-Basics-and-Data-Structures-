#1. What are semantic HTML elements? Why is using them important for web development?
  - Sementic HTML are HTML tags that have a meaning or purpose clearly defined by their name. Thee elemnts describe the role and structure of the content they contain making it easier for both browser and developers to understand the web page.

   Examples of Sementic HTML Element:
   - header: Reperent the top section or introduction of a page or section.
   - nav: Define navigation links and meanus
   - main: contain the main Content of a webpage.
   - section: Groups related content into sections.
   - article: Represent an independent piece of content.
   - aside: Contain side content.
   - footer: Represents the footer of a page or section.
   - time: Represents the dates and times.

   Why Semantic HTML is Important:

  - Improves Readability and Maintainability

  Code becomes easier to read and understand for developers.

   - Enhances SEO (Search Engine Optimization)

  Search engines better understand your content structure, improving ranking and visibility.

   - Improves Accessibility

  Screen readers and assistive technologies can interpret and navigate content correctly.

   - Provides Better Browser Compatibility

  Browsers can correctly render content according to its purpose.

   - Encourages Clean, Organized Code

  Using semantic elements gives a logical flow and reduces confusion in layout and structure.

-  1. label for Each Input

Every input field should have a label that tells users what the field is for.



```
<label for="name">Full Name:</label>
<input type="text" id="name" name="name">
```

2. Connect labels with inputs (using for and id)

Helps screen readers read the correct field name.

3. Use placeholders only as hints, not as labels

Because placeholders disappear when typing and are hard to read.

4. Add aria-label or aria-describedby when needed

Helps explain fields better for assistive technologies.

```
<input type="text" aria-label="Search here">
```
5. Ensure the form can be used with a keyboard

Users should move between fields using Tab.

Avoid disabling keyboard navigation.

6. Provide clear error messages

Tell users what they did wrong, especially visually impaired users.

Example:

```
<p id="error" role="alert">Please enter a valid email.</p>
```
7. Use proper input types (email, number, date, etc.)

Helps browsers and screen readers understand the data.

```
<input type="email">
```
8. Add fieldset & legend for grouped options

Best for radio buttons and checkboxes.

```
<fieldset>
  <legend>Choose your gender</legend>
  <input type="radio" id="male" name="gender"> Male
  <input type="radio" id="female" name="gender"> Female
</fieldset>
```
